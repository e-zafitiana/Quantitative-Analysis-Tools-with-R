### **Packages in Python for Data Analysis**

#### Packages for Data Import:
- `openpyxl`: Read, write, and modify Excel files (.xlsx).
- `pandas`: Read and manipulate Excel (xls, xlsx, csv) and other data formats.
- `pyreadstat`: Import and export files from 'SPSS', 'Stata', and 'SAS'.


#### Packages for Data Manipulation:
- `pandas`: Data manipulation (DataFrames, filtering, transformation).
- `numpy`: Numerical calculations and array manipulation.
- `scipy`: Advanced mathematical functions and statistics.


#### Packages for Data Visualization:
- `matplotlib`: Create basic plots (bar, line, scatter).
- `seaborn`: Advanced statistical visualization (heatmaps, distribution plots).
- `plotly`: Interactive and dynamic visualizations (scatter, 3D plots).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Calculation of the mode (most frequent value)
X = [24, 16, 12, 10, 12, 28, 38, 12, 28, 24]
X_mode = pd.Series(X).mode()
print("X mode:", X_mode.values)

Y = [24, 16, 12, 10, 12, 24, 38, 12, 28, 24]
Y_mode = pd.Series(Y).mode()
print("Y mode:", Y_mode.values)
```

```python
# Loading Excel Data (adjust the file path)
file_path = "bankruptcy.xlsx"
bankrupt = pd.read_excel(file_path)

# Calculating the Mean of Columns X1 and X2 (if present)
mean_X1 = bankrupt['X1'].mean()
mean_X2 = bankrupt['X2'].mean()
print("\nX1 Mean:", mean_X1)
print("X2 Mean:", mean_X2)

# Calculating Variance and Standard Deviation of Column X1
variance_X1 = bankrupt['X1'].var()
std_X1 = bankrupt['X1'].std()
print("\nX1 Variance:", variance_X1)
print("X1 Standard Deviation:", std_X1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Loading Excel Data (adjust the file path)
marital_status = pd.read_excel("MaritalStatus.xlsx")

# Ensure the data is properly loaded and formatted
if 'MaritalStatus' in marital_status.columns:
    tab = marital_status['MaritalStatus'].value_counts()
else:
    tab = marital_status.value_counts().sort_values(ascending=False)

# Converting to lists to ensure proper format
labels = [str(label) for label in tab.index]
values = list(tab.values)

# Pie Chart (Marital Status Distribution)
plt.figure(figsize=(3, 3))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Marital Status Distribution")
plt.show()
```

```python

# Bar Chart (Sorted by Frequency)
plt.figure(figsize=(5, 3))
plt.bar(labels, values, color="skyblue")
plt.title("Bar Chart - Marital Status (Sorted by Frequency)")
plt.xlabel("Marital Status")
plt.ylabel("Frequency")
plt.show()
```
