### **Graphical Data Analysis**

- Distribution shape: symmetrical or asymmetrical (positive/negative).
- Detection: outliers, trends, relationships.

**Chart Types**

``Bars Chart``: Comparison between categories. 

-	Visualizes the comparison between different categories.
-	Ideal for categorical data (e.g., product sales, survey responses).

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [4, 6, 8]

plt.figure(figsize=(3, 1.5)) 
plt.bar(categories, values)
plt.show()
```

``Linear Graph``: Trend tracking

- Displays trends over continuous data (e.g., stock prices, temperature).
- Ideal for time series analysis.

```python

plt.figure(figsize=(3, 1.5)) 

plt.plot([1, 2, 3, 4], [10, 15, 7, 12], marker='o', linestyle='-', color='green')
plt.title("Line Graph Example")
plt.show()
```

``Scatter Plot``: Explore Relationships

- Shows the relationship between two numerical variables..
- Perfect for identifying correlations (positive, negative, none).

```python

plt.figure(figsize=(3, 1.5)) 

plt.scatter([1, 2, 3, 4], [10, 20, 15, 25], color='purple')
plt.title("Scatter Plot Example")
plt.show()
```

``Pie Chart``: Show Proportions

- Visualizes the proportion of each category in a whole.
- Best for percentage distributions (e.g., market share).

```python
plt.figure(figsize=(3, 1.5)) 

plt.pie([30, 40, 30], labels=['A', 'B', 'C'], autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

``Histogram``: Display Distribution

- Visualizes the frequency distribution of a numerical dataset.
- Suitable for detecting distribution shape (normal, skewed).

```python
plt.figure(figsize=(3, 1.5)) 


data = [10, 15, 15, 20, 25, 30, 30, 30, 35, 40, 45]
plt.hist(data, bins=5, color='green')
plt.title("Histogram Example")
plt.show()
```

**Interpretation**

- Symmetric Distribution: Balanced on both sides (e.g., normal distribution).
- Positive Skew: Tail on the right (outliers with high values).
- Negative Skew: Tail on the left (outliers with low values).
- Correlation (Scatter Plot): Points form a clear pattern (positive, negative, none).
